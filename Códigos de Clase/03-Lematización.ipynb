```python
import requests

# Archivo a ser enviado
files = {'file': open('../Textos de Prueba/texto.txt', 'rb')}

# Parámetros
data = {'outf': 'tagged', 'format': 'json'}

# Enviar petición
url = "http://www.corpus.unam.mx/servicio-freeling/analyze.php"
r = requests.post(url, files=files, data=data)

# Checar qué devuelve
print("STATUS:", r.status_code)
print("HEADERS:", r.headers.get("Content-Type"))
print(r.text[:500])  # checar si realmente es JSON

# Si es JSON, parsear
try:
    obj = r.json()
    for sentence in obj:
        for word in sentence:
            print(word['token'], word['tag'], word['lemma'])
except Exception as e:
    print("No se pudo parsear como JSON:", e)
```

```python
import spacy

# Cargar modelo de español
nlp = spacy.load("es_core_news_sm")

# Texto de ejemplo
texto = "El gato duerme en la alfombra. La niña juega con su perro."

# Procesar texto
doc = nlp(texto)

# Mostrar tokens con POS
print("=== Etiquetado POS con spaCy ===")
for token in doc:
    print(f"{token.text:<15} {token.pos_:<10} {token.lemma_}")
```
