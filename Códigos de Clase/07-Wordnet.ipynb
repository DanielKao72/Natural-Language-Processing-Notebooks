```python
grafo = nx.Graph()
grafo.add_edge("a", "b")
grafo.add_edge("b", "perro")
grafo.add_edge("perro", "gato")
```

```python
nx.draw_networkx(grafo)
```

```python
todos_synset = wn.synsets("perro", lang="spa")
grafo = nx.Graph()

for cada_synset in todos_synset:
    principal = cada_synset.lemma_names("spa")
    print("Synset principal: ", principal)
    nombres_principal = principal
    hiperomimos = cada_synset.hypernyms()
    
    for cada_hiperonimo in hiperomimos:
        print("  Synset Hiperonimo: ", cada_hiperonimo.lemma_names("spa"))
        nombres_hiper = cada_hiperonimo.lemma_names("spa")
        
        for prin in nombres_principal:
            for hiper in nombres_hiper:
                grafo.add_edge(prin, hiper)
                
    hiponimos = cada_synset.hyponyms()
    
    for cada_hiponimo in hiponimos:
        print("  Synset Hiponimo: ", cada_hiponimo.lemma_names("spa"))
        nombres_hipo = cada_hiponimo.lemma_names("spa")
        
        for prin in nombres_principal:
            for hipo in nombres_hipo:
                grafo.add_edge(prin, hipo)


nx.draw_networkx(grafo)
```
