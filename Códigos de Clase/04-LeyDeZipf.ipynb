```python
print("Grfica de la distribucin de frecuencias")
distribucion = nltk.FreqDist(text1)
distribucion.plot()
```

```python
listay = []
for word, frequency in distribucion.most_common():
    listay.append(frequency)
```

```python
plt.pyplot.plot(listay)
```

```python
plt.plot(listay)
plt.xscale('log')
plt.yscale('log')
plt.show()
```
