### 1. Importamos las librerías necesarias

```python
import requests
```

### 2. Definimos la ruta y nombre del archivo donde tomará el texto

```python
FILE_NAME = "Carroll, Lewis - Alicia En El País De Las Maravillas.txt"
FILE_PATH = f"../Textos de Prueba/{FILE_NAME}"
```

### 3. Definimos el archivo y los parámetros de la salida que nos dará Freeling

```python
# Archivo a ser enviado
files = {'file': open(FILE_PATH, 'rb')}

# Parámetros
data = {'outf': 'tagged', 'format': 'json'}
```

### 4. Enviamos la petición al servicio web de Freeling

```python
# Enviar petición
url = "http://www.corpus.unam.mx/servicio-freeling/analyze.php"
response = requests.post(url, files=files, data=data)
```

### 5. Obtenemos un JSON, el cual procesamos para que se imprima únicamente los verbos en indicativo del texto

```python
# Si es JSON, parsear
try:
    obj = response.json()
    for sentence in obj:
        for word in sentence:
            tag = word['tag']
            if tag.startswith('V') and tag[2] == 'I': 
                print("Verbo:", word['token'], word['tag'], word['lemma'])
except Exception as e:
    print("No se pudo parsear como JSON:", e)
```
