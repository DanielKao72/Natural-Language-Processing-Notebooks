### 3. Importamos *wordnet*

```python
from nltk.corpus import wordnet as wn
```

### 4. Abrimos el archivo txt que contiene las palabras a buscar en *wordnet*

```python
with open("../Textos de Prueba/palabras.txt", "r", encoding="utf-8") as f:
    palabras = [line.strip() for line in f if line.strip()]
    print(palabras)
```

### 5. Creamos un diccionario con el contenido del archivo a generar.
En el diccionario se encuentra como llave la palabra y su valor es la lista de definiciones de dicha palabra. Todo obtenido de *wordnet*

```python
diccionario = {}

for palabra in palabras:
    synsets = wn.synsets(palabra)
    definiciones = [syn.definition() for syn in synsets]
    diccionario[palabra] = definiciones
```

### 6.1. Imprimimos en consola el contenido del diccionario

```python
for palabra, defs in diccionario.items():
    print(palabra)
    for i, d in enumerate(defs, 1):
        print(f"   Def {i}: {d}")
    print()
```

### 6.2. Creamos el archivo Diccionario Wordnet.txt con el contenido del diccionario generado

```python
# Guardar diccionario en un archivo de texto
with open("Diccionario Wordnet.txt", "w", encoding="utf-8") as f:
    for palabra, defs in diccionario.items():
        f.write(palabra + "\n")  # Escribe la palabra
        for i, d in enumerate(defs, 1):
            f.write(f"   Def {i}: {d}\n")  # Escribe cada definici√≥n
        f.write("\n")  # L√≠nea en blanco entre palabras

print("Archivo 'diccionario.txt' generado ‚úÖ")
```
